````python
class Solution:
    def plusOne(self, digits: List[int]) -> List[int]:
        lastdigit=len(digits)-1
        while digits[lastdigit]+1>9:
            if lastdigit==0:
                digits[lastdigit]=0
                digits.insert(0,1)
                return digits
            digits[lastdigit]=0
            lastdigit=lastdigit-1
        digits[lastdigit]=digits[lastdigit]+1
        return digits
````

### Code Explanation

#### Step 1: Initialization
```python
lastdigit = len(digits) - 1
```
- This line initializes the variable `lastdigit` to the index of the last element in the `digits` list.

#### Step 2: While Loop
```python
while digits[lastdigit] + 1 > 9:
```
- This loop continues as long as adding one to the digit at the `lastdigit` index results in a number greater than 9, which means there is a carry-over.

#### Step 3: Handling Carry-Over
```python
if lastdigit == 0:
    digits[lastdigit] = 0
    digits.insert(0, 1)
    return digits
```
- If `lastdigit` is 0 (i.e., we're at the most significant digit):
  - Set the digit at index 0 to 0 (since adding one to a 9 would roll over to 0).
  - Insert 1 at the beginning of the list (since the most significant digit increased by one due to the carry-over).
  - Return the updated list immediately, as the operation is complete.

```python
digits[lastdigit] = 0
lastdigit = lastdigit - 1
```
- If `lastdigit` is not 0:
  - Set the current digit to 0.
  - Move to the next digit to the left by decrementing `lastdigit`.

#### Step 4: Adding One without Carry-Over
```python
digits[lastdigit] = digits[lastdigit] + 1
```
- After the loop ends (when there is no more carry-over to process), add one to the current digit.

#### Step 5: Return Result
```python
return digits
```
- Return the modified list of digits.

### Time and Space Complexity

**Time Complexity**: \(O(n)\)
- In the worst case, we might have to carry over across all digits (e.g., adding one to 9999). Hence, the time complexity is proportional to the number of digits, which is \(O(n)\), where \(n\) is the length of the input list `digits`.

**Space Complexity**: \(O(1)\) (excluding the input and output)
- The function modifies the input list in place and uses a constant amount of extra space for variables like `lastdigit`. However, if we consider the space for the output list, it might require \(O(n)\) space in the case where an additional digit is inserted at the beginning (e.g., 9999 + 1 = 10000).
