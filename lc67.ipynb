````python
class Solution:
  def addBinary(self, a: str, b: str) -> str:
    s = []
    carry = 0
    i = len(a) - 1
    j = len(b) - 1

    while i >= 0 or j >= 0 or carry:
      if i >= 0:
        carry += int(a[i])
        i -= 1
      if j >= 0:
        carry += int(b[j])
        j -= 1
      s.append(str(carry % 2))
      carry //= 2

    return ''.join(reversed(s))
````

### Code Explanation

#### Step 1: Initialization
```python
s = []
carry = 0
i = len(a) - 1
j = len(b) - 1
```
- `s`: an empty list to store the result digits.
- `carry`: an integer to keep track of the carry-over during addition.
- `i`: index of the last character in string `a`.
- `j`: index of the last character in string `b`.

#### Step 2: While Loop
```python
while i >= 0 or j >= 0 or carry:
```
- This loop continues as long as there are digits to process in either `a` or `b`, or there is a non-zero carry.

#### Step 3: Adding Digits and Carry-Over
```python
if i >= 0:
  carry += int(a[i])
  i -= 1
if j >= 0:
  carry += int(b[j])
  j -= 1
```
- If `i` is non-negative, add the integer value of `a[i]` to `carry` and decrement `i`.
- If `j` is non-negative, add the integer value of `b[j]` to `carry` and decrement `j`.

#### Step 4: Calculating Result and Updating Carry
```python
s.append(str(carry % 2))
carry //= 2
```
- Append the result of `carry % 2` to `s` (this is the current digit of the result in binary).
- Update `carry` to `carry // 2` (integer division by 2), which handles the carry-over for the next iteration.

#### Step 5: Constructing the Final Result
```python
return ''.join(reversed(s))
```
- After the loop, `s` contains the binary digits of the result in reverse order, so we reverse `s` and join the elements to form the final result string.

### Time and Space Complexity

**Time Complexity**: \(O(\max(n, m))\)
- The loop iterates at most \( \max(n, m) + 1 \) times, where \( n \) is the length of `a` and \( m \) is the length of `b`. Hence, the time complexity is \(O(\max(n, m))\).

**Space Complexity**: \(O(\max(n, m))\)
- The space complexity is also \(O(\max(n, m))\) because we store the result in the list `s`, which will have at most \( \max(n, m) + 1 \) elements.
